```python
import pandas as pd
from fbprophet import  Prophet
```

```python
df = pd.read_csv('/FLOW_20211202_040000_0003.csv',encoding="shift-jis")
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

```python
df['time']
```

```python
date = df['date'][1]
time = df['time'][1]
str_datetime = f'{date} {time}'
str_datetime
```

```python
from datetime import datetime as dt
```

```python
#datetime = dt.strptime(str_datetime, '%Y/%m/%d %H:%M')
#datetime
```

```python
datetimes = []

for index, datum in df.iterrows():
    date = datum['date']
    time = datum['time']
    str_datetime = f'{date} {time}'
    datetime = dt.strptime(str_datetime, '%Y/%m/%d %H:%M')
    datetimes.append(datetime)
```

```python
datetimes
```

```python
df['ds'] = datetimes
```

```python
df.head()
```

```python
df = df.rename(columns={'flow': 'y'})
df
```

```python
model = Prophet()
```

```python
model.fit(df)
```

```python
future = model.make_future_dataframe(366,freq='d')
```

```python
future.tail()
```

```python
forecast = model.predict(future)
```

```python
import matplotlib.pyplot as plt
```

```python
model.plot(forecast)
plt.xlim(dt(2021,12,2),dt(2022,2,28))
plt.show()
```
